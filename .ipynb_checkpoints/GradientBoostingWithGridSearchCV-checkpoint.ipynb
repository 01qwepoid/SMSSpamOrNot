```python
import nltk
import re
import string
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
import pandas as pd

stopwords = nltk.corpus.stopwords.words('english')
ps = nltk.PorterStemmer()
```

```python
data = pd.read_csv('SMSSpamCollection', sep= '\t')
data.columns = ['label','body_text']
```

```python
def count_punct(text):
    count = sum([1 for char in text if char in string.punctuation])
    return round(count/(len(text) - text.count(" ")),3)*100

data['body_len'] = data['body_text'].apply(lambda x:len(x) - x.count(" "))
data['punct%'] = data['body_text'].apply(lambda x:count_punct(x))
```

```python
def clean_text(text):
    text = "".join([word.lower() for word in text if word not in string.punctuation])
    tokens = re.split('\W+',text)
    text = [ps.stem(word) for word in tokens if word not in stopwords]
    return text
```

```python
tfidf_vect = TfidfVectorizer(analyzer=clean_text)
X_tfidf = tfidf_vect.fit_transform(data['body_text'])
X_tfidf_feat = pd.concat([data['body_len'],data['punct%'], pd.DataFrame(X_tfidf.toarray())], axis =1)
```

```python
count_vect = CountVectorizer(analyzer=clean_text)
X_count = count_vect.fit_transform(data['body_text'])
X_count_feat = pd.concat([data['body_len'],data['punct%'], pd.DataFrame(X_count.toarray())], axis =1)
X_count_feat.head()
```

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import GridSearchCV
```

```python
gb = GradientBoostingClassifier()
param = {
    'n_estimators': [100,150],
    'max_depth':[7,11,15],
    'learning_rate':[0.1]
}
gs = GridSearchCV(gb, param, cv = 5, n_jobs=2)
cv_fit = gs.fit(X_tfidf_feat, data['label'])
pd.DataFrame(cv_fit.cv_results_).sort_values('mean_test_values',ascendin = False)[0:5]
```

```python
gb = GradientBoostingClassifier()
param = {
    'n_estimators': [100,150],
    'max_depth':[7,11,15],
    'learning_rate':[0.1]
}
gs = GridSearchCV(gb, param, cv = 5, n_jobs=2)
cv_fit = gs.fit(X_count_feat, data['label'])
pd.DataFrame(cv_fit.cv_results_).sort_values('mean_test_values',ascendin = False)[0:5]
```
